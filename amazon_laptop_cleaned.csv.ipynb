```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\yuva\Desktop\Amazon_Laptop_Project\amazon_laptops_raw.xls")
```

```python
df.drop(df.columns[0],axis=1,inplace=True)
```

```python
df.drop_duplicates()
```

```python
df["Price"]=df["Price"].astype(str).str.replace(",","",regex=False).str.strip()
df["Price"]=pd.to_numeric(df["Price"],errors="coerce")
df["Price"]
```

```python
df["Discount"] = (df["Discount"].astype(str).str.replace(r"[()%]", "", regex=True).str.replace("off", "", regex=False).str.strip())
df["Discount"] = pd.to_numeric(df["Discount"], errors="coerce")
df["Discount"]
```

```python
df.to_csv("amazon_laptop_cleaned.csv")
```
