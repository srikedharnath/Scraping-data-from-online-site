```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\ADMIN\OneDrive\Desktop\Amazon Scrap Data\amazon_laptop_cleaned_set.csv")
```

```python
df.drop(df.columns[0],axis=1,inplace=True)
```

```python
df
```

```python
# 1. Which brand has highest average price?

highest_avg_price=df.groupby("Brand")["Price"].mean().sort_values(ascending=False)
res=highest_avg_price.reset_index(name="Average Price")
res
```

```python
# 2. Which brand has highest average rating?

highest_avg_rating=df.groupby("Brand")["Rating"].mean().sort_values(ascending=False)
res=highest_avg_rating.reset_index(name="Average Rating")
res
```

```python
# 3. Top 5 most reviewed laptops?

most_reviewed_lap = (df.sort_values("Rating", ascending=False).head(5))
most_reviewed_lap
```

```python
# 4. What is the average discount percentage?

avg_disc_per=df.groupby("Brand")["Discount"].mean().sort_values(ascending=False).reset_index(name="Average Discount")
avg_disc_per

# Here the discount is already in percentage.
```

```python
# 5. Price vs Rating correlation value?

res=df["Price"].corr(df["Rating"])
res

# As price increases, rating slightly decreases,
# BUT the relationship is very weak.
# Because:
# Correlation is close to 0
# Not close to -1
```

```python
# 6. Distribution of laptops by RAM (8GB, 16GB, etc.)

df["RAM"] = df["RAM"].str.upper().str.strip()
distributions_on_brand=df["RAM"].value_counts().reset_index(name="Number of Laptops")
distributions_on_brand
```

```python
# 7. Which screen size is most common?

df["Screen Size"].value_counts().idxmax()
```

```python
# 8. How many laptops under ₹50,000?

res = df[df["Price"]<50000]
res.count()
```

```python
# 9. What is the average price by RAM configuration (8GB vs 16GB vs 32GB)?

res=df.groupby("RAM")["Price"].mean().reset_index(name="Average Price")
res
```

```python
# 10. What is the average price by processor type (i3, i5, Ryzen 5, etc.)?

res=df.groupby("Processor")["Price"].mean().reset_index(name="Average Price")
res
```

```python
# 11. What is the price distribution (budget, mid-range, premium)?

df["Price_Category"] = pd.cut(
    df["Price"],
    bins=[0, 40000, 70000, float("inf")],
    labels=["Budget", "Mid-Range", "Premium"]
)
df
```
